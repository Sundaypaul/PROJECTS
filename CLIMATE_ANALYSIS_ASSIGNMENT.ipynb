```python
#load the file
path = '/content/drive/MyDrive/Dataset/student_scores.csv'
df2 = pd.read_csv(path)
```

```python
df2.head()
```

```python
df2.value_counts()
```

```python
df2.nunique()
```

```python
#convert to numpy array
df2.to_numpy()
```

```python
scores = df2.drop(columns = 'Unnamed: 0')
```

```python
#1. Find the average scores for each students
average_per_student = np.mean(scores, axis=1)
df2['Average Score'] = average_per_student
print("Overall Average Score for Each Student:")
print(df2[['Unnamed: 0', 'Average Score']])
```

```python
# 2. Average Score per Course
average_per_course = np.mean(scores, axis=0)
print("\nAverage Score per Course:")
print(dict(zip(scores.columns, average_per_course)))
```

```python
# 3. Highest and Lowest Performing Students
highest_score_idx = np.argmax(average_per_student )
lowest_score_idx = np.argmin(average_per_student )

highest_performer = df2.iloc[highest_score_idx]
lowest_performer = df2.iloc[lowest_score_idx]

print("\nHighest Performing Student:")
print(f"{highest_performer['Unnamed: 0']} with an average score of {highest_performer['Average Score']}")

print("\nLowest Performing Student:")
print(f"{lowest_performer['Unnamed: 0']} with an average score of {lowest_performer['Average Score']}")
```

```python
# 4. Pass Rate (students who scored above 50 in all subjects)
pass_mask = np.all(scores > 50, axis=1)
pass_rate = np.sum(pass_mask) / len(pass_mask) * 100
print(f"\nPass Rate: {pass_rate:.2f}%")
```

```python
# 5. Variability in Courses (Standard Deviation)
std_per_course = np.std(scores, axis=0)
print("\nStandard Deviation (Variability) per Course:")
print(dict(zip(scores.columns, std_per_course)))
```
